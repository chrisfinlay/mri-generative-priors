Plug the frozen decoder into NumPyro and find the MAP latent with SVI.
Needs the `recon_model` TODO (`recon/vae_numpyro.py`, Team B) and the
operator/mask TODOs.

```python
import jax.numpy as jnp, matplotlib.pyplot as plt
from mrigen.train_vae import load_model
from mrigen.data import FastMRISlices
from mrigen.fourier import fft2c
from mrigen.masks import equispaced_mask
from mrigen.recon.vae_numpyro import reconstruct_map
from mrigen import viz, metrics
```

```python
vae = load_model('../checkpoints/vae_128.eqx', latent_dim=128)
ds = FastMRISlices('../data/processed')
x = jnp.asarray(ds[0])
mask = equispaced_mask((128, 128), acceleration=4)
y = mask * fft2c(x)
```

```python
x_map, z_map = reconstruct_map(y, mask, vae.decoder, vae.latent_dim,
                               sigma=0.01, steps=1000)
viz.panel(x, x_map); plt.show()
print('PSNR', metrics.psnr(x, x_map, data_range=1.0))
```
